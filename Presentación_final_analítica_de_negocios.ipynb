# **Naive Bayes:**


```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
nxl='/content/TravelInsurancePrediction.xlsx'
XDB=pd.read_excel(nxl,sheet_name=0)
XDB = XDB.fillna(0)

XD=np.array(XDB[['Age','Employment Type','Graduate Or Not','Annual Income In rupies','Family Members','Chronic Diseases','Frequent Flyer','Ever Travelled Abroad']])
yd=np.array(XDB['Travel Insurance'])

print("El numero de datos es:",len(XD))
ND1=len(np.where(yd==1)[0])
print("El numero de datos asegurados es:",ND1)
ND0=len(np.where(yd==0)[0])
print("El numero de datos no asegurados es:",ND0)

if len(XD) > 0:
    print("El procentaje de los asegurados es:", ND1 / len(XD))
    print("El procentaje de los no asegurados es:", ND0 / len(XD))
```

```python
mnb=GaussianNB()
mnb.fit(XD,yd)
ydp=mnb.predict(XD)

u=mnb.theta_
sigma=np.sqrt(mnb.var_)

LI=u-sigma
LS=u+sigma

print('La media de las variables por categoría:')
print(u)
print('El limite inferior de la variación de las variables por categoria es:')
print(LI)
print('El limite superior de la variación de las variables por categoria es:')
print(LS)

cm=confusion_matrix(yd,ydp)
print('La matriz de confusión es:')
print(cm)
```

```python
VN=cm[0,0]
FP=cm[0,1]
FN=cm[1,0]
VP=cm[1,1]

Exactitud=(VN+VP)/len(XD)
print(' La Exactitud es:', Exactitud)


TE=(FN+FP)/len(XD)
print('El Tasa de Error es:',TE)

Sensibilidad=VP/(VP+FP)
print('La Sensibilidad es:',Sensibilidad)

Especificidad=VN/(VN+FN)
print('La Especificidad es:',Especificidad)
```

# **Árbol de decisión**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
nxl='/content/TravelInsurancePrediction.xlsx'
XDB=pd.read_excel(nxl,sheet_name=0)
XDB.head(100)

XD=np.array(XDB[['Age','Employment Type','Graduate Or Not','Annual Income In rupies','Family Members','Chronic Diseases','Frequent Flyer','Ever Travelled Abroad']])
yd=np.array(XDB['Travel Insurance'])
```

```python
mar=DecisionTreeClassifier(criterion='gini',max_depth=4)
mar.fit(XD,yd)
ydp=mar.predict(XD)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(yd,ydp)
print(cm)

#Para evalaur el modelo usamos La matriz de confusión, y esta es: 0,0=Verdaderos negavtivos - 1,1=Verdaderos postivos - 0,1=Falsos negativos y 1,0=Falso Positivo
VN=cm[0,0]
print("Verdaderos negativos =",VN)
FP=cm[0,1]
print("Falsos positivos =",FP)
FN=cm[1,0]
print("Falsos negativos =",FN)
VP=cm[1,1]
print("Verdaderos positivos =",VP)

Ex=(VN+VP)/(VN+VP+FN+FP) #Cuantos datos o que porcentaje de estos, fueron pronosticados correctamente
print("La exactitud alcanzada por el modelo es:",Ex)
Se=VP/(FN+VP) # Sensiilidad = Como le fue con los positivos
print("La sensibilidad alcanzada por el modelo es:",Se)
Sp=VN/(VN+FP) # Especificidad = Como le fue con los negativos
print("La especificidad alcanzada por el modelo es:",Sp)
```

```python
from sklearn.tree import export_graphviz
from pydotplus import graph_from_dot_data

vs=['Age','Employment Type','Graduate Or Not','Annual Income In rupies','Family Members','Chronic Diseases','Frequent Flyer','Ever Travelled Abroad']
dot_graph=export_graphviz(mar,feature_names=vs)
graph=graph_from_dot_data(dot_graph)
graph.write_png('arbol.png')
```
